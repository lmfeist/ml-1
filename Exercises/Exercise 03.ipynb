```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
%matplotlib inline

def plot_pdf(mean, precision, n, label):
    sd = np.sqrt(1/precision)
    x = np.linspace(-8, 8, n)
    y = ss.norm.pdf(x, mean, sd)
    plt.plot(x, y, label = label)
    return y

def func(n):
    plt.figure(figsize=(12,7))

    mu = -1
    tau = 2
    mu0 = 1
    tau0 = 0.2
    D = plot_pdf(mu, tau, n, 'D')
    plot_pdf(mu0, tau0, n, 'prior')
    
    tauN = n + tau0
    muN = 1/tauN * tau * np.sum([x + tau0 * mu0 for x in D])
    
    plot_pdf(muN, tauN, n, 'posterior')
    plt.legend()
    plt.show()
```

```python
func(10)
```
