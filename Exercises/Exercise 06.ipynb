```python
%matplotlib inline

import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
import sklearn.linear_model as LM

plt.figure(figsize=(16,10))
Z = np.linspace(-5,5,100)
plt.plot(Z, st.norm.logpdf(Z), label='normal')
plt.plot(Z, st.t.logpdf(Z, df = 0.1), label='stud-t')
plt.legend()
plt.show()

X = np.linspace(-5,5,8)  

T = 2 * X + 2 * st.norm.rvs(size=(len(X)))
plt.scatter(X, T)
model = LM.Lasso(alpha = 1).fit(X.reshape(-1,1), T)
plt.plot(X, model.predict(X.reshape(-1,1)), color = 'red')

plt.show()
```
